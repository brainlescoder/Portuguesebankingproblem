```python
# Dummy variables
```

```python
obj = data.dtypes == np.object
data.columns[obj]
```

```python
len(data.columns[obj])
```

```python
dummy = pd.DataFrame()
for a in data.columns[obj]:
    dummydf = pd.get_dummies(data[a], drop_first = True)
    dummy = pd.concat([dummy, dummydf], axis= 1)
dummy
```

```python
data.shape
```

```python
data = data.drop(['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
       'month', 'day_of_week', 'poutcome', 'y'], axis = 1)
data.shape
```

```python
data = pd.concat([data, dummy], axis = 1)
data
```

```python
x = data.iloc[:, :-1]
x.shape
```

```python
y = data.iloc[:, -1]
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from pandas_profiling import ProfileReport
```

```python
confusion_matrix(y_test, lr_pred)
```

```python
print(classification_report(y_test, lr_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
rf_pred = rf.predict(x_test)
```

```python
print(classification_report(y_test,rf_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
dt_pred = dt.predict(x_test)
```

```python
print(classification_report(y_test, dt_pred))
```

```python
from sklearn.svm import LinearSVC
```

```python
svc = LinearSVC()
```

```python
svc.fit(x_train, y_train)
```

```python
svc_pred = svc.predict(x_test)
```

```python
print(classification_report(y_test, svc_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 15)
```

```python
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
```

```python
print(classification_report(y_test, knn_pred))
```

```python
err = []
for a in range(1, 40, 2):
    knn = KNeighborsClassifier(n_neighbors = a)
    knn.fit(x_train, y_train)
    knn_pred = knn.predict(x_test)
    err.append(np.mean(knn_pred != y_test))
```

```python
plt.figure(figsize = (12, 7))
plt.plot(range(1, 40, 2), err)
plt.xlabel('Kvalues')
plt.ylabel('Error')
plt.title('Optimum K value')
```

```python
from xgboost import XGBClassifier
```

```python
xg = XGBClassifier()
```

```python
xg.fit(x_train, y_train)
```

```python
xg_pred = xg.predict(x_test)
```

```python
print(classification_report(y_test, xg_pred))
```
